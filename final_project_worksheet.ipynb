```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
fraFrame = pd.read_csv('fraStats.csv', index_col=0, header=[1])
deuFrame = pd.read_csv('deuStats.csv', index_col=0, header=[1])
itaFrame = pd.read_csv('itaStats.csv', index_col=0, header=[1])
engFrame = pd.read_csv('engStats.csv', index_col=0, header=[1])
espFrame = pd.read_csv('espStats.csv', index_col=0, header=[1])
fraFrame.index.name = None; deuFrame.index.name = None; itaFrame.index.name = None
engFrame.index.name = None; espFrame.index.name = None;
fraFrame
```

```python
fig = plt.figure(figsize=(10,7)); ax = fig.subplots(); ax.set_facecolor('black')
plt.scatter(fraFrame['Tkl'],fraFrame['Int'], color='pink')
plt.scatter(deuFrame['Tkl'],deuFrame['Int'], color='yellow')
plt.scatter(itaFrame['Tkl'],itaFrame['Int'], color='lime')
plt.scatter(engFrame['Tkl'],engFrame['Int'], color='cyan')
plt.scatter(espFrame['Tkl'],espFrame['Int'], color='red')
fig.suptitle('Def. Efficiency across Top 5 Leagues in Europe 20/21 Season', fontsize=20)
plt.ylabel('Successful Interceptions During Season', fontsize=14)
plt.xlabel('Tackles Attempted During Season', fontsize=14)
legend = ['Fr. Ligue 1', 'Ger. Bundesliga 1', 'Ita. Serie A', 'Eng. Premier League', 'Spa. La Liga']
ax.legend(legend)
x = np.concatenate((fraFrame['Tkl'].array, deuFrame['Tkl'].array, itaFrame['Tkl'].array, 
                    engFrame['Tkl'].array, espFrame['Tkl'].array))
y = np.concatenate((fraFrame['Int'].array, deuFrame['Int'].array, itaFrame['Int'].array,
                  engFrame['Int'].array, espFrame['Int'].array))
m,b = np.polyfit(x,y,1); plt.plot(x, m * x + b, color='white')
plt.show()
```

```python
df = pd.read_csv('top10national.csv', index_col=0, header=[0])
df
```

```python
fig = plt.figure(figsize=(10,7)); ax = fig.subplots()
fig.suptitle('Matches won at the EURO 2020 Continental Competition', fontsize=20)
plt.ylabel('Teams', fontsize=14)
plt.xlabel('Fixtures Won Across Competition', fontsize=14)
y_pos = np.arange(10); 
ax.set_yticks(y_pos); ax.set_xticks(np.arange(7)); ax.set_yticklabels(df.index.array)
bars = ax.barh(y_pos, df['W'], align='center')
def gradient(bars):
    grad = np.atleast_2d(np.linspace(0,1,100))
    ax = bars[0].axes
    lim = ax.get_xlim()+ax.get_ylim()
    for bar in bars:
        bar.set_zorder(1)
        bar.set_facecolor("none")
        x,y = bar.get_xy()
        w, h = bar.get_width(), bar.get_height()
        ax.imshow(grad, extent=[x,x+w,y,y+h], aspect="auto", zorder=0)
    ax.axis(lim)
gradient(bars)
```
